```python
import tensorflow as tf
import numpy as np
import json
from PIL import Image

# ===========================
# Load Model
# ===========================
MODEL_PATH = "saved_model/garbage_classifier.keras"
model = tf.keras.models.load_model(MODEL_PATH)

# ===========================
# Load Class Names
# ===========================
with open("saved_model/class_names.json", "r") as f:
    class_names = json.load(f)

# ===========================
# Waste Management Information
# ===========================
waste_info = {
    "cardboard": {
        "disposal": "Flatten the cardboard and place it in the paper/cardboard recycling bin.",
        "tip": "Keep cardboard dry and free from food contamination before recycling."
    },
    "glass": {
        "disposal": "Rinse the glass item and place it in the glass recycling bin.",
        "tip": "Broken glass should be wrapped carefully before disposal for safety."
    },
    "metal": {
        "disposal": "Clean metal cans or containers and place them in the metal recycling bin.",
        "tip": "Recycling metal saves energy and reduces mining."
    },
    "paper": {
        "disposal": "Dispose of clean paper in the paper recycling bin.",
        "tip": "Wet or oily paper should not be recycled."
    },
    "plastic": {
        "disposal": "Rinse plastic containers and place them in the plastic recycling bin.",
        "tip": "Reducing single-use plastics helps protect the environment."
    },
    "trash": {
        "disposal": "Dispose of non-recyclable waste in the general waste bin.",
        "tip": "Separate recyclable items whenever possible to reduce landfill waste."
    }
}


# ===========================
# Image Preprocessing
# ===========================
def preprocess_image(image):

    image = image.convert("RGB")
    image = image.resize((224, 224))

    img_array = np.array(image, dtype=np.float32)

    img_array = np.expand_dims(img_array, axis=0)

    return img_array


# ===========================
# Prediction Function
# ===========================
def predict_image(image):

    processed = preprocess_image(image)

    prediction = model.predict(processed, verbose=0)

    predicted_index = np.argmax(prediction)

    predicted_class = class_names[predicted_index]

    confidence = float(np.max(prediction) * 100)

    disposal = waste_info[predicted_class]["disposal"]

    tip = waste_info[predicted_class]["tip"]

    return {
        "class": predicted_class,
        "confidence": confidence,
        "disposal": disposal,
        "tip": tip
    }
```
